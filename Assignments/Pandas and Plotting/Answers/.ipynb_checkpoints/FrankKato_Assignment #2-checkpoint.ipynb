```python
import numpy as np
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python

#A function that creates a random dataset
# with numerical and categorical features
def create_dataset(n_samples, n_features, n_classes):
    #set the seed
    np.random.seed(42)
    X = np.random.randn(n_samples, n_features)
    y = np.random.randint(0, n_classes, n_samples)
    #combing the numerical and categorical features
    z = np.hstack((X, y.reshape(-1, 1)))
    # add random column names to the dataset
    z = pd.DataFrame(z, columns=['col' + str(i) for i in range(n_features + 1)])
    # add a column with random strings combinations
    z['string'] = np.random.choice(['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'], n_samples)
    # return the dataset
    return z
```

```python
# create two datasets and join them based on one common variable
print("firstDataset")
firstDataset=create_dataset(5,3,2) #first data set
print(firstDataset)

print("secondDataset")
secondDataset=create_dataset(5,4,2) #second data set
print(secondDataset)

print("jointDataset")
jointDataset=pd.merge(firstDataset, secondDataset, on='col1', how='left', suffixes=['_l','_r']) #joint data set based on col1
print(jointDataset)
```

This joins datasets based on col0 where all values for the left data set(firstDataset) are retained and NaN fills missing values in the right data set(secondDataset)
